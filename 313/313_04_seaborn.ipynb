## seaborn함수

```python
# 사전 라이브러리 설치 : pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 팁 데이터 불러오기
tips = sns.load_dataset("tips")

print(tips.head()) # 팁 데이터가 표형태로 출력됨

# 팁 금액 분포 시각화

# sns.distplot(tips['tip'], kde=True) # 사라진 함수라 함
"""
Please adapt your code to use either `displot` (a figure-level function with
similar flexibility) or `histplot` (an axes-level function for histograms).
"""

sns.histplot(tips['tip'], kde=True) # kde=True 선형도를 같이 출력함. false로 하면 선 그래프가 숨겨짐
plt.show()
```

```python
# 성별에 따른 팁 금액 비교 (박스플롯)
sns.boxplot(x='sex', y='tip', data=tips)
# I 형 선과 박스, 동그라미 형태의 도형들로 이루어진 그래프
# I 형은 데이터의 범위
# 박스의 바닥은 데이터의 하위 25%
# 박스의 중간 선은 평균값
# 박스의 위는 데이터의 상위 25%
# I 바깥 동그라미는 이상수치
plt.show()
```

```python
# 팁 금액과 총 지출액 간의 관계 (산점도)
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
# 데이터 불러오기
df = pd.read_csv('dataset/diabetes.csv')

# 데이터 정보 확인
print(df.info())

# 1. Violinplot: Outcome에 따른 Glucose 분포 시각화
plt.figure(figsize=(8, 6))
sns.violinplot(x='Outcome', y='Glucose', data=df, palette='muted')
plt.title('Violin Plot of Glucose by Outcome')
plt.xlabel('Outcome')
plt.ylabel('Glucose Level')
plt.show()
```

```python
# Pairplot: 주요 변수들 간의 관계 시각화
plt.figure(figsize=(10, 10))
sns.pairplot(df, vars=['Glucose', 'BMI', 'Age'], hue='Outcome', palette='coolwarm')
plt.suptitle('Pairplot of Glucose, BMI, and Age (Colored by Outcome)', y=1.02)
plt.show()
```

```python
#  Lmplot: Age와 BMI 간의 회귀 관계 시각화
sns.lmplot(x='Age', y='BMI', data=df, hue='Outcome', palette='Set2', height=6, aspect=1.5)
plt.title('Linear Regression Plot of Age vs. BMI by Outcome')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.show()
```
