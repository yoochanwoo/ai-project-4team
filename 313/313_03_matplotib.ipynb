## matplotlib 함수

```python
# 사전 라이브러리 설치 : pip install matplotlib
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.linspace(0, 2*3.14, 100) # linspace(시작값, 끝값, 곡률) x축을 만듦 
y = np.sin(x) # y값 지정 함수. numpy에는 삼각함수가 내장되어있음

# 선 그래프
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine Wave')
plt.show()
```

```python
# 산점도
plt.scatter(x, y)
plt.show()
```

```python
# 막대 그래프
data = [3, 8, 1, 10]
plt.bar(['A', 'B', 'C', 'D'], data)
plt.show()
```

```python
# 히스토그램
np.random.seed(19680801)
data = np.random.randn(1000)
plt.hist(data, bins=30)
plt.show()
```
