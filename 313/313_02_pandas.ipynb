```python
# 사전설치 : pip install pandas (터미널창 단축키 : ctrl + shift + `)
import pandas as pd

# 데이터프레임 생성
data = {'이름': ['Alice', 'Bob', 'Charlie'], '나이': [25, 30, 22], '성별': ['여', '남', '남']}
df = pd.DataFrame(data, index=['A', 'B', 'C'])  # 행 인덱스를 지정
# 딕셔너리 데이터를 배열형태로 만듦

str1 = "홍길동"
print(df)
print(type(df)) # 변수타입이 특이함
# pandas에서 제공하는 테이블 형태의 변수타입

print(type(str1))
```

```python
# data = {'이름': ['Alice', 'Bob', 'Charlie'], '나이': [25, 30, 22], '성별': ['여', '남', '남']}
# df = pd.DataFrame(data, index=['A', 'B', 'C'])

# 첫 번째 행 선택 (인덱스 0)
first_row = df.iloc[0]
print(first_row)

first_row2 = df.iloc[0,:] # 행선택은 열지정을 생략할 수 있음
print(first_row2)
# 열지정(,:)을 생략하지 않는 것이 규칙적이라서 외우기 쉽다

# 두 번째 열 (나이) 선택
age_column = df.iloc[:, 1]
print(age_column)

# 첫 번째 행부터 두 번째 행까지, 모든 열 선택
subset = df.iloc[0:2, :] # 인덱스 규칙에 주의
print(subset)

subset2 = df.iloc[0:2] # 행선택은 열지정을 생략할 수 있음
print(subset2)
```

```python
# iloc 은 인덱스값으로 접근하고,
# loc 은 키값으로 접근함!

"""
행 키값 : [A,B,C]
열 키값 : [이름,나이,성별]
"""

# '이름' 열 선택
name_column = df.loc[:, '이름']
print(name_column)

# 'A' 행 선택
first_row = df.loc['A']
print(first_row)

# 나이가 25 이상인 행 선택
selected_rows = df.loc[df['나이'] >= 25]
print(selected_rows)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 당뇨병 데이터 불러오기 (데이터 파일 경로 변경 필요)
df = pd.read_csv('dataset/diabetes.csv')

df.head()    # 앞에서 5개 항목 출력   cf. tail() : 뒤에서 5개 항목
```

```python
df.tail() # head() 와 tail() 을 같이 쓸 수는 없음
```

```python
df.info() # 파일의 정보를 알려줌
```

```python
df.describe() # 각 열의 값에 통계를 내줌
```

```python
# 각 변수의 분포 확인
df.hist(figsize=(10, 10)) # figsize=(길이, 높이) : 사이즈의 크기를 지정함
plt.show() # 시각화 함수
```

```python
# 변수 간 상관관계 시각화
#print(df.corr()) # df.corr() : 각 열의 값들끼리의 비율을 계산

sns.heatmap(df.corr(), annot=True) # annot=True 수치를 출력함. False입력시 숫자 없이 색깔로만 표현
plt.show()
# 계산한 값을 시각화하여 출력
```

```python
# 결측치 확인 : null값을 찾음
df.isnull().sum()
```

```python
# 결측치 평균값으로 채우기
df.fillna(df.mean(), inplace=True)  # inplace = True 는 변경된 설정으로 덮어쓴다는 의미
print(df)
# 파일을 직접 수정하는 것은 아님
```
