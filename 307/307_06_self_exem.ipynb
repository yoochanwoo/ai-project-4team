## 숫자야구

```python
import random

rnum1 = random.randint(1,9)
rnum2 = random.randint(0,9)
while rnum1 == rnum2:
    rnum2 = random.randint(0,9)
rnum3 = random.randint(0,9)
while rnum1 == rnum3 or rnum2 == rnum3:
    rnum3 = random.randint(0,9)
rnum4 = random.randint(0,9)
while rnum1 == rnum4 or rnum2 == rnum4 or rnum3 == rnum4:
    rnum4 = random.randint(0,9)
ball = 0
strike = 0
#print(rnum1, rnum2, rnum3, rnum4)
cnt=1
while True:
    print(f"{cnt}회차")
    if cnt==9:
        print("마지막 기회")
    
    user_num=input("4자리의 수를 입력하세요 : ")
    while len(user_num)!=4 or not user_num.isnumeric():
        print("4자리 수가 아닙니다")
        user_num=input("다시 4자리의 수를 입력하세요 : ")

    unum1 = int(user_num[0])
    unum2 = int(user_num[1])
    unum3 = int(user_num[2])
    unum4 = int(user_num[3])
    print(unum1, unum2, unum3, unum4)
    
    if rnum1!=unum1:
        if rnum1==unum2:
            ball+=1
        if rnum1==unum3:
            ball+=1
        if rnum1==unum4:
            ball+=1
    else:
        strike+=1
            
    if rnum2!=unum2:
        if rnum2==unum1:
            ball+=1
        if rnum2==unum3:
            ball+=1
        if rnum2==unum4:
            ball+=1
    else:
        strike+=1
        
    if rnum3!=unum3:
        if rnum3==unum1:
            ball+=1
        if rnum3==unum2:
            ball+=1
        if rnum3==unum4:
            ball+=1
    else:
        strike+=1
        
    if rnum4!=unum4:
        if rnum4==unum2:
            ball+=1
        if rnum4==unum3:
            ball+=1
        if rnum4==unum1:
            ball+=1
    else:
        strike+=1

    if rnum1==unum1 and rnum2==unum2 and rnum3==unum3 and rnum4==unum4:
        print(f"{cnt}번만에 성공!")
        break
    elif ball==0 and strike==0:
        print("아웃")
    else:
        print(f"{strike}스트라이크 {ball}볼")
    ball = 0
    strike = 0
    cnt+=1
    
    if cnt==10:
        print("실패")
        break
```
